## Importing Libraries of Interest

```python
# Importing Select Python Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Clicking the play button on the left will run this entire cell.

# To run a highlighted line, press ⌘/Cntrl + Shift + Enter!
```

We can now use functions from `pandas`, `matplotlib.pyplot` and `seaborn` within our Jupyter Notebook.

Let's begin by loading a `.csv` file from the [Github repository](https://github.com/MJAlexander/social_media_workshop) associated with Monica Alexander's [Social Media Workshop](https://mjalexander.github.io/social_media_workshop/). To do so, we'll be using the `read_csv` function from `pandas`.

```python
# Reading data into our notebook:

df = pd.read_csv("https://raw.githubusercontent.com/MJAlexander/social_media_workshop/main/data/country_indicators.csv")

# We can take a look at the first view observations using head() — just like in base R:

df.head()
```

## A Quick Plot via `seaborn`

```python
# Generating the same basic plot we produced via reticulate:

sns.set_theme()

sns.scatterplot(x = "life_expectancy", y = "tfr", hue = "gdp", data = df)

# To display plots in non-Jupyter contexts, it helps to use functions from matplotlib.pyplot:

plt.show()
```

## Saving a File to Google Drive

How can we save the plot we created to Google Drive? 

**Note**: If you're playing around with this notebook on your local machine, you can jump to Step 3.

### Step 1

Run the following lines of code on Google Colab:

```python
from google.colab import drive
drive.mount('/drive')
```

Create a folder (e.g., named Python) within Google Drive.

### Step 3

We can save our image using the `savefig` function from `matplotlib`:

```python
# In Jupyter, we'll have to generate/save the figure in the same cell!

sns.scatterplot(x = "life_expectancy", y = "tfr", hue = "gdp", data = df)

plt.savefig('/drive/My Drive/Python/fig1.png', dpi = 300)

# If you're running Jupyter from Anaconda, you can run the following line:

# plt.savefig('fig1.png', dpi = 300)
```

To save our data frame, we can turn to `pandas`:

```python
# Let's use the to_csv function:

df.to_csv('/drive/My Drive/Python/country_ind.csv')

# If you're running this notebook locally, you can run the following line:

# df.to_csv("country_ind.csv")
```
